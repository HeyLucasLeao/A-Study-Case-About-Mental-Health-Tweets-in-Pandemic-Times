```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import re
from os import listdir
```

```python
PATH = r'C:\Users\heylu\Documents\github\A Study Case About Mental Health Tweets in Pandemic Times\data'
```

```python
df = pd.DataFrame()
for f in listdir(PATH + '\\' + '2018'):
    dataset_from_files_in_path = pd.read_csv(PATH + '\\' + '2018' + '\\' + f)
    df = pd.concat([df, dataset_from_files_in_path])
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df
```

```python
frequency = df
frequency['quantidade'] = 1
frequency = frequency.groupby('date').sum()[['quantidade']]
fig = px.bar(frequency)
fig
```

```python
frequency.index = pd.to_datetime(frequency.index)
frequency_per_month = frequency.groupby(pd.Grouper(freq='M')).sum()
fig = px.bar(frequency_per_month)
fig
```

```python
frequency.reset_index(inplace=True)
frequency['dia_da_semana'] = frequency['date'].dt.day_name()
dias_traduzidos = {'Monday': 'Segunda', 
                   'Tuesday': 'Ter√ßa',
                  'Wednesday': 'Quarta',
                  'Thursday': 'Quinta',
                  'Friday': 'Sexta',
                  'Saturday': 'S√°bado',
                  'Sunday': 'Domingo'}
frequency['dia_da_semana'] = frequency['dia_da_semana'].apply(lambda x: dias_traduzidos[x])
dia_da_semana = frequency.groupby('dia_da_semana').sum()
fig = px.bar(dia_da_semana)
fig
```

```python
fig = px.line(data_frame=frequency, y='quantidade', x='date')
fig
```
